```python
import nltk
import pandas as pd
```

```python
nltk.download('punkt')
```

```python
fake=pd.read_csv("fake.csv")
fake.drop(fake.iloc[:, 4:172], inplace=True, axis=1)
fake
```

```python
true=pd.read_csv("true.csv")
true
```

```python
fake['genuene']=0
true['genuene']=1
df=pd.concat([fake,true],axis=0)
df
```

```python
df=df.reset_index(drop=True)
df=df.drop(['title','subject','date'],axis=1)
```

Data Preprocessing

```python
#tokenize
from nltk.tokenize import word_tokenize
df['text']=df['text'].apply(word_tokenize)
```

```python
#stemming
from nltk.stem import SnowballStemmer as ss 
sb=ss("english",ignore_stopwords=False)
def stem(text):
    return [sb.stem(i) for i in text]
df['text']=df['text'].apply(stem)
```

```python
#stop words
def stop_words(text):
    return [i for i in text if len(i)>2]
```

```python
df['text']=df['text'].apply(" ".join)
df
```

Splitting Data

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(df['text'],df['genuene'],test_size=0.25)
x_train
```

Vectorization

```python
#vectorization

from sklearn.feature_extraction.text import TfidfVectorizer
tfdf=TfidfVectorizer(max_df=0.7)

tfdf_train=tfdf.fit_transform(x_train)
tfdf_test=tfdf.transform(x_test)
```

ML Model

```python
# ML moedel

from sklearn.linear_model import LogisticRegression
m1=LogisticRegression(max_iter=900)
m1.fit(tfdf_train,y_train)

predict1=m1.predict(tfdf_test)
 
```

```python
from sklearn.metrics import accuracy_score
scr1=accuracy_score(y_test,predict1)
scr1
```

```python
from sklearn.linear_model import PassiveAggressiveClassifier
m2=PassiveAggressiveClassifier(max_iter=100)
m2.fit(tfdf_train,y_train)

pred2=m2.predict(tfdf_test)
scr2=accuracy_score(y_test,pred2)
scr2
```
